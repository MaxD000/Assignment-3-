```python
#assignment3
import random

def load(): #function to load words
    words = {}
    with open("words.txt", 'r') as file:
        for word in file:
            word = word.upper()  
            length = len(word)
            if length not in words:
                words[length] = []
            words[length].append(word)  
    return words

def select(words, word_length):#function du select a random word in the list 'words.txt'
    return random.choice(words[word_length])

def play():#function to play the game
    word_length = int(input("Write a word length: "))
    words = load()
    solution = select(words, word_length)
    attempt = 7
    display = "_ " * len(solution)
    letters = ""
    score = 0

    while attempt > 0:
        print("Guess the word: ", display)
        proposition = input("Guess a letter: ").upper()

        if proposition in letters:
            print("You already guessed this letter")
            continue

        letters += proposition

        if proposition in solution:
            print("Good guess")
        else:
            attempt -= 1
            print(f"Wrong guess!:{attempt}")

        display = ""
        for x in solution: 
            if x in letters:
                display += x
                score+=10
            else:
                display += "_ "
                score-=5

        if "_" not in display:
            print(f"You won! The word was: {solution}")
            break

    if attempt == 0:
        print(f"Game over. The word was: {solution}")

    print("End of the game, score:", score)

    p=input("Do you want to play again ? :")
    if p=="yes":
        if __name__ == "__main__":
            play() 
    else:
        print ("End of the game")

if __name__ == "__main__":
    play()
```
